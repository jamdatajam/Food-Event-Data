```python
###STEP TWO: Add a plot; we are going to try a regression plot here
#start with the code for the subplots and backdrop
fig, axes = plt.subplots(2,2, figsize=(15,10)) #figsize = the output size of the plot
plt.subplots_adjust(wspace = .3, hspace = .3) #wspace/hspace = the amount of empty space as spacings

#Now add the actual plot. I like to use seaborn for regression stuff
ax = sns.regplot(x = 'Vendors', y = 'Visitors', data = ftfN2, ax = axes[0,0])
ax.set(ylim=(-1000,8000))
```

```python
#There you go! Looks good!
#we can see that there is a best fit line in the middle, the scattered points that represent
#the actual visitor vs vendors data
#and a shaded area around the regression line to show how much error there is in the predicted fit line
```

```python
#STEP TWO, CONTINUED...
#So what about the other blank plots?
# well let's fill'em in!

#PLOT (0,0) Position 1
#start with the code for the subplots and backdrop
fig, axes = plt.subplots(2,2, figsize=(15,10)) #figsize = the output size of the plot
plt.subplots_adjust(wspace = .3, hspace = .3) #wspace/hspace = the amount of empty space as spacings

#Now add the actual plot. I like to use seaborn for regression stuff
ax = sns.regplot(x = 'Vendors', y = 'Visitors', data = ftfN2, ax = axes[0,0])
ax.set(ylim=(-1000,8000))

#PLOT (0,1) Position 2
#So I want to do something involving stars (social media review scores) and visitors
#and if possible, include the number of reviews in some meaningful way
ax = ftfN2.plot.scatter(x = 'Stars', y = 'Visitors', c = ftfN2['Reviews'], s = ftfN2['Reviews'],
                       ax = axes[0,1], cmap = 'Oranges')
ax.set(ylim=(-500, 7000), xlim = (2,5.25))
#Where c is the color of the markers, s = size of markers, and cmap is the color pattern I am using
```

```python
#As you can see, the color intensity increases as the number of reviews increases
#So too does the size.
#Feasibly, you could have color intensity and size be different metrics, or columns in the data
#For example:
#PLOT (0,0) Position 1
#start with the code for the subplots and backdrop
fig, axes = plt.subplots(2,2, figsize=(15,10)) #figsize = the output size of the plot
plt.subplots_adjust(wspace = .3, hspace = .3) #wspace/hspace = the amount of empty space as spacings

#Now add the actual plot. I like to use seaborn for regression stuff
ax = sns.regplot(x = 'Vendors', y = 'Visitors', data = ftfN2, ax = axes[0,0])
ax.set(ylim=(-1000,8000))

#PLOT (0,1) Position 2
#So I want to do something involving stars (social media review scores) and visitors
#and if possible, include the number of reviews in some meaningful way
ax = ftfN2.plot.scatter(x = 'Stars', y = 'Visitors', c = ftfN2['Reviews'], s = ftfN2['Entry']*50,
                       ax = axes[0,1], cmap = 'Oranges')
ax.set(ylim=(-500, 7000), xlim = (2,5.25))
```

```python
#Now let's put some more plots in those two bottom positions
#I like contour-ish-esque plots, so let's do that next

#PLOT (0,0) Position 1
#start with the code for the subplots and backdrop
fig, axes = plt.subplots(2,2, figsize=(15,10)) #figsize = the output size of the plot
plt.subplots_adjust(wspace = .3, hspace = .3) #wspace/hspace = the amount of empty space as spacings

#Now add the actual plot. I like to use seaborn for regression stuff
ax = sns.regplot(x = 'Vendors', y = 'Visitors', data = ftfN2, ax = axes[0,0])
ax.set(ylim=(-1000,8000))

#PLOT (0,1) Position 2
#So I want to do something involving stars (social media review scores) and visitors
#and if possible, include the number of reviews in some meaningful way
ax = ftfN2.plot.scatter(x = 'Stars', y = 'Visitors', c = ftfN2['Reviews'], s = ftfN2['Entry']*50,
                       ax = axes[0,1], cmap = 'Oranges')
ax.set(ylim=(-500, 7000), xlim = (2,5.25))

#PLOT (1,0) Position 3
ax = sns.kdeplot(ftfN2.Month, ftfN2.Visitors, ax = axes[1,0], cmap = 'rainbow')
ax.set(ylim=(-500,9000), xlim = (0.5,13))
```

```python
#And position 4:
#PLOT (0,0) Position 1
#start with the code for the subplots and backdrop
fig, axes = plt.subplots(2,2, figsize=(15,10)) #figsize = the output size of the plot
plt.subplots_adjust(wspace = .3, hspace = .3) #wspace/hspace = the amount of empty space as spacings

#Now add the actual plot. I like to use seaborn for regression stuff
ax = sns.regplot(x = 'Vendors', y = 'Visitors', data = ftfN2, ax = axes[0,0])
ax.set(ylim=(-1000,8000))

#PLOT (0,1) Position 2
#So I want to do something involving stars (social media review scores) and visitors
#and if possible, include the number of reviews in some meaningful way
ax = ftfN2.plot.scatter(x = 'Stars', y = 'Visitors', c = ftfN2['Reviews'], s = ftfN2['Entry']*50,
                       ax = axes[0,1], cmap = 'Oranges')
ax.set(ylim=(-500, 7000), xlim = (2,5.25))

#PLOT (1,0) Position 3
ax = sns.kdeplot(ftfN2.Month, ftfN2.Visitors, ax = axes[1,0], cmap = 'rainbow')
ax.set(ylim=(-500,9000), xlim = (0.5,13))

#PLOT (1,1) Position 4
ax = sns.kdeplot(ftfN2.Start_Time, ftfN2.Visitors, ax = axes[1,1], cmap = 'rainbow')
ax.set(ylim=(-1000,9000), xlim = (900, 2000))
```

```python
#These two plots show "hotspots" where the most data is congregated
#It looks like the best month for the most visitors is sometime between July and September
#And the best starttime might be around 11:00 AM (the times are in Military Time)
```
